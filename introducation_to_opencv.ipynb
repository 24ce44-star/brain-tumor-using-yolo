```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
#load the image
image = cv2.imread('/content/opharshit.jpg', cv2.IMREAD_COLOR)
cv2_imshow(image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
inverted_gray_image = cv2.bitwise_not(gray_image)
gray_image
```

```python
mask = image[:, :, 0] == 0
filtered_image = image.copy()
filtered_image[~mask] = 0  # Set all other pixels to black
cv2_imshow(filtered_image)
```

```python
cv2_imshow(mask.astype(np.uint8) * 255)  # Convert mask to 8-bit and display
```

```python
type(image)
```

```python
#invert the greyscale image
inverted_image = cv2.bitwise_not(image)
```

```python
scaled_original = cv2.resize(image, (64, 64), interpolation=cv2.INTER_NEAREST)
scaled_inverted = cv2.resize(inverted_image, (64, 64), interpolation=cv2.INTER_NEAREST)
```

```python
# save the result
cv2.imwrite('output_image.jpg', inverted_image)
```
